```python
from statsmodels.tsa.stattools import adfuller

def adf_test(dataset):
     """
    Perform the Augmented Dickey-Fuller (ADF) test to check for stationarity.

    Parameters:
    - dataset (pandas.Series or array-like): The time series data to be tested for stationarity.

    Returns:
    None

    Prints the following ADF test results:
    1. ADF Statistic: The Augmented Dickey-Fuller test statistic.
    2. P-Value: The p-value obtained from the test.
    3. Num Of Lags: The number of lags used in the regression when determining the test statistic.
    4. Num Of Observations Used For ADF Regression: The number of observations used in the ADF regression.
    5. Critical Values: The critical values for the test at different confidence levels.

    Note:
    - This function uses the 'AIC' criterion to automatically select the number of lags.
     """
     dftest = adfuller(dataset, autolag = 'AIC')
     print("1. ADF : ",dftest[0])
     print("2. P-Value : ", dftest[1])
     print("3. Num Of Lags : ", dftest[2])
     print("4. Num Of Observations Used For ADF Regression:",      dftest[3])
     print("5. Critical Values :")
     for key, val in dftest[4].items():
         print("\t",key, ": ", val)

adf_test(amazon_data['Close'])
```

```python
from scipy.stats import boxcox

# Apply Box-Cox transformation to the closing share price
amazon_data['BoxCox_Close'], lambda_boxcox = boxcox(amazon_data['Close'])

# Difference the transformed data to make it stationary
amazon_data['diff_BoxCox_Close'] = amazon_data['BoxCox_Close'].diff().fillna(0)

# Plot the differenced Box-Cox transformed data
amazon_data['diff_BoxCox_Close'].plot(title='Differenced Box-Cox Transformed Amazon Closing Share Price', figsize=(10, 6))
plt.ylabel('Differenced Box-Cox Close')
plt.xlabel('Date')
plt.show()

# Check stationarity again with ADF test
result_diff_log = adfuller(amazon_data['diff_BoxCox_Close'].dropna())
print('ADF Statistic (Differenced Log):', result_diff_log[0])
print('p-value:', result_diff_log[1])
print('Critical Values:')
for key, value in result_diff_log[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
print(amazon_data.shape)
train=amazon_data.iloc[:-30]
test=amazon_data.iloc[-30:]
print(train.shape,test.shape)
```

```python
# Test model performance

# Create and fit the best SARIMA model based on manual approach
best_model = SARIMAX(amazon_data['Close'], order=(2, 1, 2))
best_model_fit = best_model.fit()

# Display the summary of the best model
print("Best Model Summary:")
print(best_model_fit.summary())

# Plot diagnostics for the best model
best_model_fit.plot_diagnostics()
plt.show()
```

```python
start=len(train)
end=len(train)+len(test)-1
pred=best_model_fit.predict(start=start,end=end,typ='levels').rename('SARIMA Predictions')
pred.plot(legend=True)
test['Close'].plot(legend=True)
```
